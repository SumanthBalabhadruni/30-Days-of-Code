```python
def coinchange(tamount,coins):
  amount=tamount
  coins=sorted(coins)
  new=[]
  i=len(coins) - 1
  new.append(coins[i])
  while amount > 0:
    if amount >= coins[i]:
      amount = amount - coins[i]
      new.append(coins[i])
    if amount < coins[i]:
      i=i-1
  return new
coins=[1,2,5,10,20,50,100,1000]
k=coinchange(2035,coins)
print('minimum no of coins required to make total amount is {} and the corresponding coins are {}'.format(len(k),k))
```

```python
tot=50 # no of kgs
items=[[20,100],[30,120],[10,60]]
for i in range(len(items)):
  items[i].append(items[i][-1]/items[i][0])
items.sort(key=lambda x:x[2], reverse=True)
val=0
new=[]
for i in range(len(items)):
  if tot==0:
    break
  if tot > items[i][0]:
    tot-=items[i][0]
    val+=items[i][1]
    new.append(items[i][1])
  else:
    val+=(tot/items[i][0])*items[i][-2]
    new.append((tot/items[i][0])*items[i][-2])
print(val)
```
