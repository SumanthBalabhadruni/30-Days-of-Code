# Dynamic Programming

## Top Down Memorization

```python
def fibmemo(n,memo):
  if n==1:
    return 0
  if n==2:
    return 1
  if not n in memo:
    memo[n]=fibmemo(n-1,memo) + fibmemo(n-2,memo)
  return memo[n]
memo={}
print(fibmemo(6,memo))
```

## Buttom Up Tabulation

```python
def fibtab(n):
  tb=[0,1]
  for i in range(2,n):
    tb.append(tb[i-1] + tb[i-2])
  return tb[-1]
print(fibtab(10))
```
